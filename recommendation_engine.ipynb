```python
#import librairies
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
books=pd.read_csv('bestsellers with categories.csv')
```

```python
books.head()
```

```python
books.info()
```

```python
books.isnull().sum()
```

```python
#getting unique rows/Title
books=books.drop_duplicates(subset=['Name'], ignore_index=True)
```

```python
books
```

```python
books.reset_index(inplace=True)
books = books.rename(columns = {'index':'book_id'})
```

```python
books
```

```python
columns=['Name','Author', 'Genre','Reviews']
```

```python
def combine_features(data):
    features=[]
    for i in range(0, data.shape[0]):
        features.append(data['Name'][i]+' '+data['Author'][i]+' '+data['Genre'][i])
    return features
```

```python
books['combined_features']=combine_features(books)
```

```python
books
```

```python
#Convert the text from the nex column to a matrix word counts
cm = CountVectorizer().fit_transform(books['combined_features'])
```

```python
cs = cosine_similarity(cm)
```

```python
print(cs)
```

```python
#Get the title of the book the reader likes
title = books['Name'][1]
title
```

```python
#book written by Stephen King
```

```python
book_id = books[books.Name == title]['book_id'].values[0]
```

```python
book_id
```

```python
scores = list(enumerate(cs[book_id]))
```

```python
print(scores)
```

```python
sorted_scores = sorted(scores, key= lambda x:x[1], reverse=True)
```

```python
sorted_scores = sorted_scores[1:]
```

```python
sorted_scores
```

```python
# create a loop to print the first 5 books from the sorted list
j=0
print('The five most recommended books to '+title+' are:\n')
for item in sorted_scores:
    book_title = books[books.book_id == item[0]]['Name'].values[0]
    print (j+1, book_title)
    j=j+1
    if j >= 5:
        break
```
